```python
# Write a program to work out how many days you have lived for.
from datetime import date
birthday = date(int(input("year: ")),int(input("month: ")),int(input("day: ")))
today = date.today()
print("days lived: ", str(today - birthday).split(",")[0])
```

```python
# Write a program to work out how many seconds you have lived for.
from datetime import date, datetime
birthday = datetime(int(input("year: ")),int(input("month: ")),int(input("day: ")))
today = datetime.now()
difference = today - birthday
print("seconds lived: ", round(difference.total_seconds()))
```

```python
# Make a game of rock, paper, scissors against the computer.
# Algorithm
# Tell user to enter either rock,paper or scissors
# Get the response
# Generate a random number from 1 to 3 (1=rock,2=paper,3=scissors)
# Compare user selection and computer selection
# Display who wins.
from random import randint
choices = ["rock", "paper", "scissors", "quit"]
computer_choice = randint(1,3)
user_choice = input("Enter rock, paper or scissors: ")
user_choice = choices.index(user_choice.lower()) + 1
print("Computer chose", choices[computer_choice -1])
if user_choice == computer_choice:
    print("Draw")
elif user_choice-computer_choice == 1 or (user_choice == "rock" and computer_choice == "scissors"):
    print("You win!")
else:
    print("Computer Wins!")
```

```python
# For the rock, paper, scissors game above, ensure that the user enters a valid 
# entry. Add a loop structure to play several times and keep a running score.
i = 1
wins = 0
losses = 0
while True:
    print("Round 1:")
    print("Wins:", wins, "Losses:", losses)
    user_choice = input("Enter 'rock', 'paper' or 'scissors'. Enter 'quit' to exit. ")
    while user_choice.lower() not in choices:
        print("Invalid choice, try again.")
        user_choice = input("Enter 'rock', 'paper' or 'scissors'. Enter 'quit' to exit. ")
    if user_choice.lower() == "quit":
        break
    user_choice = choices.index(user_choice.lower()) + 1
    print("Computer chose", choices[computer_choice -1])
    if user_choice == computer_choice:
        print("Draw")
    elif user_choice-computer_choice == 1 or (user_choice == "rock" and computer_choice == "scissors"):
        print("You win!")
        wins += 1
    else:
        print("Computer Wins!")
        losses += 1
```

```python
# Write a program that will give the answer to logic gate questions e.g.
# Enter logic gate : OR
# Enter first input : 1
# Enter second input :0
# Result = 1
# It should work for the logic gates OR, XOR, NAND and NOR gates.
gate = input("Enter logic gate: ")
x = input("Enter first input: ")
y = input("Enter second input: ")
flip = [1, 0]
if gate.upper() == "OR":
    print(x or y)
if gate.upper() == "AND":
    print(x and y)
if gate.upper() == "XOR":
    if (not x) != (not y):
        print(1)
    else:
        print(0)
if gate.upper() == "NAND":
    print(flip[int(x and y)])
if gate.upper() == "NOR":
    print(flip[int(x or y)])
```

```python
# Write a program that will display all the factors of an input number that
# are bigger than 1.
# (e.g. the factors of the number 12 are 6,4,3 and 2 because they divide
# into 12 exactly).
# Hint: To find out whether a number X is a factor of Y use:
# If Y mod X == 0 (there is nothing remaining when Y is divided by X)
n = int(input("Enter a positive whole number: "))
print(sorted(list(x for tup in ([i, n//i] for i in range(1, int(n**0.5)+1) if n % i == 0) for x in tup)))
```
