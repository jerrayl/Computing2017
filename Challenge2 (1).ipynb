```python
''' 
Write a program for a game where the computer generates a
random starting number between 20 and 30.
The player and the computer can remove 1,2 or 3 from the number
in turns. Something like this:

Starting number: 25
How many do you want to remove? 3
22 left
Computer removes 2
20 left
...

The player who has to remove the last value to bring the number
down to 0 is the loser.

1 left
Computer removes 1
You win!

Easy: Get the computer to choose a number between 1â3 at random. 
'''
from random import randint
number = randint(20,30)
print("Starting number: ", number)
while number>0:
    remove = int(input("How many do you want to remove? "))
    number -= remove
    print(number, "left")
    if number == 0:
        print("You lose!")
        break
    elif number < 3:
        comremove = randint(1, number)
    else:
        comremove = randint(1, 3)
    print("Computer removes", comremove)
    number -= comremove
    print(number, "left")
    if number == 0:
        print("You Win!")
    
```

```python
'''
Write a program for a game where the computer generates a
random starting number between 20 and 30.
The player and the computer can remove 1,2 or 3 from the number
in turns. Something like this:

Starting number: 25
How many do you want to remove? 3
22 left
Computer removes 2
20 left
...

The player who has to remove the last value to bring the number
down to 0 is the loser.

1 left
Computer removes 1
You win!

Hard: Get the computer to employ a strategy to try and win. 
'''
from random import randint
number = randint(20,30)
print("Starting number: ", number)
while number>0:
    remove = int(input("How many do you want to remove? "))
    number -= remove
    print(number, "left")
    if number == 0:
        print("You lose!")
        break
    comremove = randint(1,3)
    if number == 1 or number == 2 or number - comremove == 3 or number - comremove == 2:
        comremove = 1
    if number < 4:
        comremove = number - 1
    if number == 1:
        print("Computer removes 1")
        print("0 left")
        print("You Win!")
        break
    print("Computer removes", comremove)
    number -= comremove
    print(number, "left")
```

```python
'''
Write a program for a Higher / Lower guessing game.
The computer randomly generates a sequence of 10 numbers
between 1 and 13. The player after seeing each number in turn
has to decide whether the next number is higher or lower.
If you get 9 guesses right you win the game.

Starting number : 12
Higher(H) or lower(L)? L
Next number 8
Higher(H) or lower(L)? L
Next number 11
You lose

Hints
- Use a condition controlled loop (do until, while etc) to control the
game.
- You do not need to remember all 10 numbers, but  just the current and
the next number.
- Remember to keep a count of the number of guesses. 
'''
from random import randint
sequence = []
for _ in range(10):
    sequence.append(randint(1,13))
print("Starting number:", sequence[0])
for i in range(0,9):
    if i==9:
        print("You win")
    else:
        hl = input("Higher(H) or lower(L)?")
        print("Next number:", sequence[i+1])
        if hl == "H" and sequence[i] > sequence[i+1]:
            print("You lose")
            break
        elif hl == "L" and sequence[i] < sequence[i+1]:
            print("You lose")
            break
        
```

```python
'''
Extend the above Higher / Lower number guessing game by
- Giving the player two lives
- Allowing only H or L to be entered 
'''
from random import randint
sequence = []
have_lives = True
for _ in range(10):
    sequence.append(randint(1,13))
print("Starting number:", sequence[0])
for i in range(0,9):
    if i==9:
        print("You win")
    else:
        hl = input("Higher(H) or lower(L)?")
        print("Next number:", sequence[i+1])
        if (hl == "H" and sequence[i] > sequence[i+1]) or (hl == "L" and sequence[i] < sequence[i+1]):
            if have_lives == False:
                print("You lose")
                break
            else:
                have_lives = False
                print("Wrong! One life left.")
```

```python
'''
Write a program to count the number of words in a sentence.
The user enters a sentence.
The program outputs the number of words in the sentence.
Hint: Look for spaces and full stops in the string. 
'''
sentence = input("Enter a sentence: ")
if sentence.count(".") != 1:
    print("Not a valid sentence.")
else:
    print(len(sentence.split(" ")))
```
